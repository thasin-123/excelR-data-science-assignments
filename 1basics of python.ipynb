#### 1. Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself. 

```python
num = int(input("Enter a number: "))
if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

#### 2. Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message. 

```python
import random

# Generating two random numbers
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

user_answer = int(input(f"What is {num1} * {num2} ? "))

correct_answer = num1 * num2

if user_answer == correct_answer:
    print("Correct!, Well done.")
else:
    print("Wrong answer.")
    print("The correct answer is:", correct_answer)
```

#### 4. Problem Statement (Question): Hospital Billing System 
A hospital charges its patients based on the following: 

Room charges per day: ₹2,000 

Doctor consultation fee (flat): ₹1,500 

Lab test charges: Based on the number of tests taken. Each test costs ₹300 

Medicine charges: Total cost of medicines provided 

Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied 

#### Write a Python program to calculate the final bill for a patient who: 

Stayed for 4 days 

Had 3 lab tests 

Had medicine charges of ₹2,400 

```python
# hospital charges

room_charge = 2000
doctor_consultation_fee = 1500
lab_test_charge = 300

n1 = int(input("total days of room stay :"))
n2 = int(input("total number of lab tests :"))

total_room_charges = n1 * room_charge
total_lab_test_charges = n2 * lab_test_charge
medicine_charge = int(input("enter total medicine amount :"))

print("\n----- Hospital Bill -----")
print( "Room stay amount:" ,total_room_charges)
print( "Lab charges:", total_lab_test_charges)
print( "Medicine amount:" , medicine_charge)
print( "Doctor fee:" , doctor_consultation_fee)

total_bill = total_room_charges + total_lab_test_charges + medicine_charge + doctor_consultation_fee
print("Total amount : ", total_bill)


if total_bill > 10000:
    discount = (total_bill * 10)/100
    final_bill = total_bill - discount
    print("Discount:", discount)
    print("Final bill to pay", final_bill)
else:
    final_bill = total_bill
    print("Final bill", final_bill)
    
    
```
