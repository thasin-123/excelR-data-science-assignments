```python

X = data.drop('income', axis=1)  # all features
y = data['income']   # target
```

```python
# identifying numerical and categorical columns

num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object']).columns
print(num_cols)
print(cat_cols)
```

```python
# scaling the numerical data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
# X[num_cols]
```

### 2. Encoding Techniques: 

```python
# encoding categorical data

# One-hot for small categories
one_hot_cols = [col for col in cat_cols if X[col].nunique() < 5]

# Label encoding for large categories
label_cols = [col for col in cat_cols if X[col].nunique() >= 5]

# One-hot
X = pd.get_dummies(X, columns=one_hot_cols, drop_first=True)

# Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in label_cols:
    X[col] = le.fit_transform(X[col])
```

```python
# encoding target column

le_y = LabelEncoder()
y = le_y.fit_transform(y)
```

```python
data1 = pd.DataFrame(X, columns= data.columns)
```

```python
y = pd.DataFrame(y, columns=['income>50k'])
```

```python
data1 = X.join(y)
```

### 3. Feature Engineering: 

```python
# skewness

num_cols = ['age','fnlwgt','education_num','capital_gain',
                  'capital_loss','hours_per_week']
# num_cols = data.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    print(col,':', data1[col].skew())
```

```python
# features 1
data1['total_capital'] = data['capital_gain'] - data['capital_loss']
```

```python
# features 2
data1['hours_per_age'] = data['hours_per_week'] / data['age']
```

```python
# skewness

real_num_cols = ['age','fnlwgt','education_num','capital_gain',
                 'capital_loss','hours_per_week',
                 'total_capital','hours_per_age']

for col in real_num_cols:
    print(col, data1[col].skew())
```

```python
# log function

data1['capital_gain_log'] = np.log1p(data['capital_gain'])

data1['capital_loss_log'] = np.log1p(data['capital_loss'])

data1['total_capital_log'] = np.log1p(data1['total_capital'])

data1['fnlwgt_log'] = np.log1p(data['fnlwgt'])
```

```python
# droping unwanted columns

data1.drop(['capital_gain', 'capital_loss', 'total_capital'], axis=1, inplace=True)

data1.drop('fnlwgt', axis=1, inplace=True)
```
