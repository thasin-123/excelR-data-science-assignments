### Section C: Exploring Pandas 

#### 11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then: 
- Use .info(), .describe() 
- Select 'Name' and 'City' columns 
- Drop 'City' column 
- Fill any missing values in 'Salary' column with the mean 
- Remove any duplicate rows 

```python
import numpy as np
import pandas as pd


data = {
    "Name": ["Rahul", "Anita", "Kiran", "Sneha", "Arjun",
             "Meena", "Vikram", "Divya", "Ravi", "Priya"],
    "Age": [25, 30, 28, 22, 35, 27, 31, 22, 29, 26],
    "City": ["Delhi", "Mumbai", "Chennai", "Hyderabad", "Bangaluru",
             "Amaravati", "Kolkata", "Mysuru", "Tirupati", "Vizag"],
    "Salary": [50000, 60000, 55000, 40000, 75000,
               np.nan, 68000, 45000, 58000, 49000]
}
df = pd.DataFrame(data)
df
```

```python
# using info

df.info()
```

```python
# using describe

df.describe()
```

```python
# finding null values

df.isna().sum()
```

```python
# select name and city

df[['Name','City']]
```

```python
# droping city

df.drop('City', axis=1, inplace=True)
df
```

```python
# filling mean with null values

df['Salary'] = df['Salary']. fillna(df['Salary'].mean())
df
```

```python
# removing duplicate rows

df.drop_duplicates(subset=["Age"], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```
