```python
import numpy as np
import pandas as pd
```

```python
# loading data
patient = pd.read_csv("patient_Data.csv")
patient
```

```python
patient.shape
```

```python
# info of dataset
patient.info()
```

```python
#finding nulls
patient.isna().sum()
```

```python
# statistics
patient.describe()
```

```python
# droping unwanted columns
patient.drop(['ReceptionistID',	'CheckInTime'] , axis=1,inplace=True)
```

```python
# grouping the data by dept
patient.groupby(['Department' , 'BillAmount']).sum()
```

```python
# droping the duplicates 
patient = patient.drop_duplicates(subset='Patient_ID')
patient
```

```python
# filling the null values with mean 
patient['BillAmount'] = patient['BillAmount'].fillna(patient['BillAmount'].mean())
```

```python
patient
```

```python
# loding billing dataset
billing = pd.read_csv('Billing_Data.csv')
billing
```

```python
# merging the two data set on basis of Patient_ID 
final_result = pd.merge(patient, billing, on='PatientID')
final_result
```
