```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
```

#### Task 1 :  state hypothesis testing

#### Task 2. Calculate the Test Statistic: 

```python
sample_mean = 3050
alpha = 0.05

# Known population values
pop_mean = 4000
pop_std = 125 

n = 25

z_test = (sample_mean - pop_mean)/(pop_std/np.sqrt(n))
p_value = 1 - stats.norm.cdf(z_test)

print(f"Z-test: {z_test:.3f}")
print(f"P-value: {p_value:.3f}")

```

#### Task 3. Determine the Probability and compare: 

```python

if p_value < alpha:
    print("Reject the null hypothesis. There is sufficient evidence that the weekly operating cost has increased and the old cost model may not be accurate.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to say that the weekly operating cost has increased.")
```

# Estimation And Confidence Intervals 

#### a. Build 99% Confidence Interval Using Sample Standard Deviation 

```python

# Data
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48,
        1.20, 1.33, 1.18, 1.22, 1.29]

# Sample size
n = len(data)

# Sample mean
mean = np.mean(data)

# Sample standard deviation
std = np.std(data, ddof=1)

# Standard error
se = std / np.sqrt(n)

# Confidence level
confidence = 0.99

# t critical value
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Margin of error
margin_error = t_critical * se

# Confidence interval
lower = mean - margin_error
upper = mean + margin_error

print("Sample Mean:", mean)
print("Sample Standard Deviation:", std)
print("t Critical Value:", t_critical)
print("Margin of Error:", margin_error)
print("99% Confidence Interval:", (lower, upper))
```

#### b. Build 99% Confidence Interval Using Known Population Standard Deviation 

```python
from scipy.stats import norm

# Data
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 
        1.48, 1.20, 1.33, 1.18, 1.22, 1.29]

# Sample mean
mean = np.mean(data)

# Known population standard deviation
sigma = 0.2

# Sample size
n = len(data)

# Z critical value for 99% confidence
z = norm.ppf(0.995)

# Standard Error
se = sigma / np.sqrt(n)

# Margin of Error
me = z * se

# Confidence Interval
lower = mean - me
upper = mean + me

print("Sample Mean:", mean)
print("99% Confidence Interval:", (lower, upper))
```
