```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

#### 1. Data Cleaning and Preparation:

```python
# load dataset
data = pd.read_csv('Cardiotocographic.csv') 
data
```

```python
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
# filling the null values

data['LB']= data['LB'].fillna(data['LB'].mean())
data['AC']= data['AC'].fillna(data['AC'].median())
data['DS']= data['DS'].fillna(data['DS'].mean())
data['DP']= data['DP'].fillna(data['DP'].mean())
data['MLTV']= data['MLTV'].fillna(data['MLTV'].median())
data['Width']= data['Width'].fillna(data['Width'].median())
data['Tendency']= data['Tendency'].fillna(data['Tendency'].mode()[0])
data['NSP']= data['NSP'].fillna(data['NSP'].mode()[0])
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
# box plot
sns.catplot(data,
           kind='box',
           height=4,
           aspect=1.5)
plt.xticks(rotation=45)
plt.show()
```

```python
# deleting the outliers

cols = ['LB','ASTV','MSTV','MLTV','ALTV','Width']

data1 = data.copy()

for col in cols:
    Q1 = data1[col].quantile(0.25)
    Q3 = data1[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    data1 = data1[(data1[col] >= lower) & (data1[col] <= upper)]
```

```python
data1.head()
```

```python
# box plot
sns.catplot(data1,
           kind='box',
           height=4,
           aspect=1.5)
plt.xticks(rotation=45)
plt.show()
```

```python
# inter quantile range 
cols = ['LB','ASTV','MSTV','MLTV','ALTV','Width']

data1 = data.copy()

for col in cols:
    Q1 = data1[col].quantile(0.25)
    Q3 = data1[col].quantile(0.75)
    IQR = Q3 - Q1
    print(f"{col}-IQR = {IQR}")
```

#### 3. Data Visualization: 

```python
# histogram plot

cols=['LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV','Width']

for col in cols:
    sns.histplot(data1,x=col, kde =True, color='black', edgecolor='black')
    plt.show()

```

```python
#bar plot
sns.catplot(data1,x='Tendency',y='LB', kind='bar',height=10,aspect=2)
plt.xticks(rotation=90)
plt.show()

```
